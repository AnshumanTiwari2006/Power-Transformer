```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PowerTransformer

```

```python
df = pd.read_csv('./Dataset/crop_yield_dataset.csv')
```

```python
df.sample(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Yield (tonnes/ha)', axis = 1),
                                                   df['Yield (tonnes/ha)'],
                                                   test_size = 0.2,
                                                   random_state = 42)
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

r2_score(y_test, y_pred)
```

```python
lr = LinearRegression()
np.mean(cross_val_score(lr, df.drop('Yield (tonnes/ha)', axis = 1), df['Yield (tonnes/ha)'], scoring = 'r2'))
```

```python
for x in X_train.columns:
    plt.figure(figsize=(14, 4))

# PDF / Histogram with KDE
    plt.subplot(1, 2, 1)
    sns.histplot(X_train[x], kde=True, color='skyblue', bins=30)
    plt.title('Histogram with KDE')

# QQ Plot
    plt.subplot(1, 2, 2)
    stats.probplot(X_train[x], dist="norm", plot=plt)
    plt.title('QQ Plot')

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
for x in X_train_scaled.columns:
    plt.figure(figsize=(14, 4))

# PDF / Histogram with KDE
    plt.subplot(1, 2, 1)
    sns.histplot(X_train_scaled[x], kde=True, color='skyblue', bins=30)
    plt.title('Histogram with KDE')

# QQ Plot
    plt.subplot(1, 2, 2)
    stats.probplot(X_train_scaled[x], dist="norm", plot=plt)
    plt.title('QQ Plot')

    plt.tight_layout()
    plt.show()
```

```python
pt = PowerTransformer(method = 'box-cox')

X_train_transformed = pt.fit_transform(X_train + 0.000001)
X_test_transformed = pt.transform(X_test + 0.000001)

pd.DataFrame({'cols' : X_train.columns, 'box_cox_lambdas': pt.lambdas_})
```

```python
lr = LinearRegression()

lr.fit(X_train_transformed, y_train)

y_pred2 = lr.predict(X_test_transformed)

r2_score(y_test, y_pred2)
```
